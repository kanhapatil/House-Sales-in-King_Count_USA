# Dealing with Missing Values in Python


## How to deal with missing values
- **Drop the missing values**
    - drop the variable
    - drop the data entry


- **Replacing the missing values**
    - replace it with an average
    - replace it by frequency
    - replace it based on other functions
    
 
**Question 1: how would you deal with missing values for categorical data**
**Ans :** 1) Replace missing values with mode, 2) Replace missing values with most offen.

### How to drop missing values in pandas
- dataframe.dropna(axis = 0)
    - axis = 0 ( drop the complete row )
    - axis = 1 ( drop the complete column )
    
- dataframe.dropna(subset = ["Price"], axis = 0, inplace = True)


### How to replace missing values in python
- dataframe.replace(missing_vales, new_values)

# Data Normalization in Python
- **Uniform the features value with different range.**





### Simple feature scaling
- df["length"] = df["length"] / df["length"].max()

### Min Max scaling
- df["length"] = (df["length"] - df["length"].min()) / (df["length"].max() - df["length"].min())

### Z-score
- df["lenght"] = (df["length"] - df["length"].mean()) / df["length"].std()

# Binning in Python
- Grouping of values into "bins"
- Convert numeric into categorical variables
- Group a set of numerical values into a set of "bins"
- "Price" is a feature range from **5000 to 45500** (in order to have **better representation** of price)


- **Prices :**
- 5000, 10000, 12000 --- considerd as  **Low**
- 30000, 32000 ---------- considerd as  **Mid**
- 40000, 45000, 50000 -- considerd as  **High**

**Code**
- bins = np.linspace(min(df["price"]), max(df["price"]), 4)
- group_names = ["low", "median", "high"]
- df["price-binned"] = pd.cut(df["price"], bins, labels = group_names, include_lowest = True)
